```python
import os, glob, random, torch, numpy as np
from tqdm import tqdm
from transformers import RobertaTokenizer, RobertaForSequenceClassification, get_scheduler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, f1_score, roc_auc_score
from sklearn.linear_model import LogisticRegression
from sklearn.feature_extraction.text import TfidfVectorizer
from xgboost import XGBClassifier
import nlpaug.augmenter.word as naw
from nlpaug.augmenter.word import SynonymAug
import nltk

# Reproducibility
random.seed(42); np.random.seed(42); torch.manual_seed(42); torch.cuda.manual_seed_all(42)
os.environ["PYTHONHASHSEED"] = str(42)

# NLTK Resources
nltk.download('wordnet')
nltk.download('omw-1.4')
nltk.download('averaged_perceptron_tagger')

# Hyperparameters
hyperparams = {
    "lr": 1e-5, "weight_decay": 0.01, "epochs": 20, "batch_size": 16,
    "max_length": 512, "early_stopping_patience": 2, "tfidf_max_features": 5000
}

train_path = "/kaggle/input/ptsd-new-multimodel-dataset/sorted_multimodal_data/text/training"
val_path = "/kaggle/input/ptsd-new-multimodel-dataset/sorted_multimodal_data/text/validation"

# Load data
def load_text_data(folder_path):
    texts, labels = [], []
    for label_name in ['PTSD', 'NO PTSD']:
        label = 1 if label_name == 'PTSD' else 0
        for file in glob.glob(os.path.join(folder_path, label_name, '*.txt')):
            try:
                with open(file, 'r', encoding='utf-8') as f:
                    texts.append(f.read().strip())
                    labels.append(label)
            except: continue
    return texts, labels

# Define augmenters
swap_aug = naw.RandomWordAug(action='swap', aug_p=0.1)
delete_aug = naw.RandomWordAug(action='delete', aug_p=0.1)
augmenters = [swap_aug, delete_aug]

def augment_texts(texts, augmenters):
    augmented = []
    for text in texts:
        augmenter = random.choice(augmenters)
        try:
            aug_text = augmenter.augment(text)
        except Exception as e:
            print(f"‚ö†Ô∏è Augmentation failed for: {text[:30]}... | {str(e)}")
            aug_text = text
        if isinstance(aug_text, list):
            aug_text = " ".join(aug_text)
        augmented.append(aug_text)
    return augmented

# Load + Split + Augment
print("üì• Loading data...")
train_texts, train_labels = load_text_data(train_path)
val_texts, val_labels = load_text_data(val_path)
all_texts = train_texts + val_texts
all_labels = train_labels + val_labels
train_texts, temp_texts, train_labels, temp_labels = train_test_split(all_texts, all_labels, test_size=0.4, stratify=all_labels, random_state=42)
val_texts, test_texts, val_labels, test_labels = train_test_split(temp_texts, temp_labels, test_size=0.5, stratify=temp_labels, random_state=42)
aug_texts = augment_texts(train_texts, augmenters)
train_texts += aug_texts
train_labels += train_labels.copy()

# TF-IDF + LR
vectorizer = TfidfVectorizer(max_features=hyperparams["tfidf_max_features"])
X_train_tfidf = vectorizer.fit_transform(train_texts)
X_val_tfidf = vectorizer.transform(val_texts)
X_test_tfidf = vectorizer.transform(test_texts)

lr_model = LogisticRegression(max_iter=1000)
lr_model.fit(X_train_tfidf, train_labels)

# Roberta Model
tokenizer = RobertaTokenizer.from_pretrained("roberta-base")
model = RobertaForSequenceClassification.from_pretrained("roberta-base", num_labels=2)
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Freeze lower Roberta layers
for name, param in model.named_parameters():
    if any(f"encoder.layer.{i}" in name for i in range(4)):
        param.requires_grad = False
model.to(DEVICE)

# Encode text data
def encode_data(texts, labels):
    enc = tokenizer(texts, truncation=True, padding=True, max_length=hyperparams["max_length"], return_tensors="pt")
    return torch.utils.data.TensorDataset(enc["input_ids"], enc["attention_mask"], torch.tensor(labels))

train_dataset = encode_data(train_texts, train_labels)
val_dataset = encode_data(val_texts, val_labels)
test_dataset = encode_data(test_texts, test_labels)
train_loader = torch.utils.data.DataLoader(train_dataset, batch_size=hyperparams["batch_size"], shuffle=True)
val_loader = torch.utils.data.DataLoader(val_dataset, batch_size=hyperparams["batch_size"])
test_loader = torch.utils.data.DataLoader(test_dataset, batch_size=hyperparams["batch_size"])

# Optimizer & Scheduler
optimizer = torch.optim.AdamW(filter(lambda p: p.requires_grad, model.parameters()), lr=hyperparams["lr"], weight_decay=hyperparams["weight_decay"])
scheduler = get_scheduler("linear", optimizer, 0, len(train_loader)*hyperparams["epochs"])

import matplotlib.pyplot as plt

# Track metrics
train_accuracies, val_accuracies = [], []
train_losses, val_losses = [], []

best_val_loss = float('inf'); counter = 0
for epoch in range(1, hyperparams["epochs"] + 1):
    model.train(); train_loss, train_correct, train_total = 0.0, 0, 0
    for batch in tqdm(train_loader, desc=f"Epoch {epoch}"):
        input_ids, attention_mask, labels = [b.to(DEVICE) for b in batch]
        outputs = model(input_ids=input_ids, attention_mask=attention_mask, labels=labels)
        loss = outputs.loss; logits = outputs.logits
        optimizer.zero_grad(); loss.backward(); optimizer.step(); scheduler.step()
        train_loss += loss.item()
        preds = torch.argmax(logits, dim=1)
        train_correct += (preds == labels).sum().item(); train_total += labels.size(0)

    avg_train_loss = train_loss / len(train_loader)
    train_acc = train_correct / train_total
    train_losses.append(avg_train_loss)
    train_accuracies.append(train_acc)

    # Validation
    model.eval(); val_loss, val_correct, val_total = 0.0, 0, 0
    with torch.no_grad():
        for batch in val_loader:
            input_ids, attention_mask, labels = [b.to(DEVICE) for b in batch]
            outputs = model(input_ids=input_ids, attention_mask=attention_mask, labels=labels)
            val_loss += outputs.loss.item()
            preds = torch.argmax(outputs.logits, dim=1)
            val_correct += (preds == labels).sum().item(); val_total += labels.size(0)

    avg_val_loss = val_loss / len(val_loader)
    val_acc = val_correct / val_total
    val_losses.append(avg_val_loss)
    val_accuracies.append(val_acc)

    print(f"\nüìä Epoch {epoch} | üîπ Train Loss: {avg_train_loss:.4f} | Train Acc: {train_acc*100:.2f}% | üîπ Val Loss: {avg_val_loss:.4f} | Val Acc: {val_acc*100:.2f}%")

    if avg_val_loss < best_val_loss:
        best_val_loss = avg_val_loss; counter = 0
        torch.save(model.state_dict(), "best_roberta.pt")

# Plot after training
plt.figure(figsize=(10, 5))
plt.plot(train_accuracies, label='Train Accuracy')
plt.plot(val_accuracies, label='Validation Accuracy')
plt.title("üìà Accuracy per Epoch")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 5))
plt.plot(train_losses, label='Train Loss')
plt.plot(val_losses, label='Validation Loss')
plt.title("üìâ Loss per Epoch")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.grid(True)
plt.show()

# Predict Roberta probabilities
def get_roberta_probs(dataloader):
    model.eval(); probs = []
    with torch.no_grad():
        for batch in dataloader:
            input_ids, attention_mask, _ = [b.to(DEVICE) for b in batch]
            logits = model(input_ids=input_ids, attention_mask=attention_mask).logits
            probs.extend(torch.softmax(logits, dim=1)[:, 1].cpu().numpy())
    return np.array(probs)

# Ensemble Evaluation
model.load_state_dict(torch.load("best_roberta.pt"))
val_roberta_probs = get_roberta_probs(val_loader)
test_roberta_probs = get_roberta_probs(test_loader)
val_tfidf_probs = lr_model.predict_proba(X_val_tfidf)[:, 1]
test_tfidf_probs = lr_model.predict_proba(X_test_tfidf)[:, 1]

meta_val_X = np.vstack([val_roberta_probs, val_tfidf_probs]).T
meta_test_X = np.vstack([test_roberta_probs, test_tfidf_probs]).T
meta_y_val = np.array(val_labels)
meta_y_test = np.array(test_labels)

xgb_meta = XGBClassifier(n_estimators=100, max_depth=4, use_label_encoder=False, eval_metric='logloss')
xgb_meta.fit(meta_val_X, meta_y_val)

ensemble_preds = xgb_meta.predict(meta_test_X)
ensemble_probs = xgb_meta.predict_proba(meta_test_X)[:, 1]

print("\nüì¢ ENSEMBLE TEST RESULTS (Roberta + TFIDF ‚Üí XGBoost)")
print(f"‚úÖ Accuracy : {accuracy_score(meta_y_test, ensemble_preds)*100:.2f}%")
print(f"‚úÖ F1 Score : {f1_score(meta_y_test, ensemble_preds)*100:.2f}%")
print(f"‚úÖ ROC AUC  : {roc_auc_score(meta_y_test, ensemble_probs)*100:.2f}%")
```
