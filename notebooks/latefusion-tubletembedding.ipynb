```python
!pip install pytorchvideo
# import libraires

import random
import numpy as np
import joblib
import os, re, torch, torch.nn as nn
from torch.utils.data import Dataset, DataLoader
from torch.optim import AdamW
from torchvision import transforms, models
from PIL import Image
from collections import Counter
from tqdm import tqdm
from pytorchvideo.models.resnet import create_resnet
from transformers import BertForSequenceClassification, BertTokenizer
from transformers.optimization import get_cosine_schedule_with_warmup
from xgboost import XGBClassifier


# ‚îÄ‚îÄ‚îÄ Reproducibility: Fix Random Seeds ‚îÄ‚îÄ‚îÄ
import torch
SEED = 42
os.environ["PYTHONHASHSEED"] = str(SEED)
random.seed(SEED)
np.random.seed(SEED)
torch.manual_seed(SEED)
if torch.cuda.is_available():
    torch.cuda.manual_seed(SEED)
    torch.cuda.manual_seed_all(SEED)
torch.backends.cudnn.deterministic = True
torch.backends.cudnn.benchmark = False

print(f"All seeds set to {SEED}.")

# (2) tunable settings
device        = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
NUM_CLASSES   = 2
SEQ_LEN       = 50
EPOCHS        = 20
BATCH_SIZE    = 8
LR            = 8.5e-5        
MAX_TEXT_LEN  = 128
VIDEO_DEPTH   = 50             # 50 

root_video_dir  = '/kaggle/input/ptsd-new-multimodel-dataset/sorted_multimodal_data/video/training'
val_session_dir = '/kaggle/input/ptsd-new-multimodel-dataset/sorted_multimodal_data/video/validation'
train_audio_dir = '/kaggle/input/ptsd-new-multimodel-dataset/sorted_multimodal_data/audio/training'
val_audio_dir   = '/kaggle/input/ptsd-new-multimodel-dataset/sorted_multimodal_data/audio/validation'
train_text_dir = "/kaggle/input/ptsd-new-multimodel-dataset/sorted_multimodal_data/text/training"
val_text_dir   = "/kaggle/input/ptsd-new-multimodel-dataset/sorted_multimodal_data/text/validation"



ckpt_video = '/kaggle/input/ptsd-uni-models/pytorch/default/1/best_tublett_embedding_model.pth'
ckpt_audio = '/kaggle/input/ptsd-uni-models/pytorch/default/1/best_effnet_vit_ensemble.pth'
ckpt_text  = '/kaggle/input/ptsd-uni-models/pytorch/default/1/ensemble_model.pth'

class_names = ["NO PTSD", "PTSD"]
label_map   = {c: i for i, c in enumerate(class_names)}

# (3)  helper utilities

def strip_prefix(sd, pref):
    return { (k[len(pref):] if k.startswith(pref) else k): v for k,v in sd.items() }

def build_entries(text_root):
    entries=[]
    for cls in class_names:
        d=os.path.join(text_root, cls)
        entries += [{'id': os.path.splitext(f)[0], 'label': label_map[cls]}
                    for f in os.listdir(d) if f.endswith('.txt')]
    return entries

# (4) data augmentation + custom Dataset

vid_tf = transforms.Compose([
    transforms.RandomResizedCrop(224, scale=(0.8,1.0)),
    transforms.RandomHorizontalFlip(),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406],
                         std=[0.229, 0.224, 0.225])
])

aud_tf = transforms.Compose([
    transforms.RandomApply([transforms.ColorJitter(0.2,0.2)], p=0.5),
    transforms.Resize(224), transforms.CenterCrop(224), transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406],
                         std=[0.229, 0.224, 0.225])
])

tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')

class PTSDDataset(Dataset):
    def __init__(self, entries, sessions, vroot, aroot, troot):
        self.e, self.s = entries, sessions
        self.vroot, self.aroot, self.troot = vroot, aroot, troot

    def __len__(self):
        return len(self.e)

    def __getitem__(self, idx):
        sid, lbl = self.e[idx]['id'], self.e[idx]['label']

        # ‚îÄ‚îÄ Video frames (jaisa pehle tha) ‚îÄ‚îÄ
        frames = []
        for sess in self.s:
            d = os.path.join(self.vroot, sess, class_names[lbl])
            if os.path.isdir(d):
                frames += [os.path.join(d,f) for f in sorted(os.listdir(d))
                           if f.startswith(sid) and f.endswith('.png')]
        if not frames:
            vid = torch.zeros(3, SEQ_LEN, 224, 224)
        else:
            if len(frames) < SEQ_LEN:
                frames += frames[-1:] * (SEQ_LEN - len(frames))
            frames = frames[:SEQ_LEN]
            vid = torch.stack([vid_tf(Image.open(p).convert('RGB')) for p in frames], 1)

        # ‚îÄ‚îÄ Audio spectrograms: SEQ_LEN frames ‚îÄ‚îÄ
        # Aap ke paas multiple spectrograms per patient: sid*.png
        a_dir = os.path.join(self.aroot, class_names[lbl])
        # List all matching files
        aud_files = []
        if os.path.isdir(a_dir):
            aud_files = [os.path.join(a_dir, f)
                         for f in sorted(os.listdir(a_dir))
                         if f.startswith(sid) and f.endswith('.png')]
        # Agar files nahi milti, to placeholder frame use karo
        if not aud_files:
            base_img = Image.new('RGB', (224,224))
            aud_files = [None]  # marker for blank

        # Pad/truncate list to SEQ_LEN
        if len(aud_files) < SEQ_LEN:
            aud_files += aud_files[-1:] * (SEQ_LEN - len(aud_files))
        aud_files = aud_files[:SEQ_LEN]

        # Load & transform each spectrogram
        aud_frames = []
        for fpath in aud_files:
            if fpath is None or not os.path.exists(fpath):
                img = Image.new('RGB', (224,224))
            else:
                img = Image.open(fpath).convert('RGB')
            aud_frames.append(aud_tf(img))
        # Stack into shape (3, SEQ_LEN, 224, 224)
        aud = torch.stack(aud_frames, dim=1)

        # ‚îÄ‚îÄ Text transcript (unchanged) ‚îÄ‚îÄ
        t_path = os.path.join(self.troot, class_names[lbl], f'{sid}.txt')
        text = open(t_path).read().strip() if os.path.exists(t_path) else ''
        enc  = tokenizer(text, return_tensors='pt',
                         padding='max_length', truncation=True,
                         max_length=MAX_TEXT_LEN)
        ids, mask = enc['input_ids'][0], enc['attention_mask'][0]

        return vid, aud, ids, mask, lbl

# (5) train / val splits + DataLoader 

# entries
train_e, val_e = build_entries(train_text_dir), build_entries(val_text_dir)

# class weights
cnts = Counter([e['label'] for e in train_e])
tot = sum(cnts.values())
class_wts = torch.tensor([tot/(NUM_CLASSES * cnts[i]) for i in range(NUM_CLASSES)], device=device)

# sessions for training (under root_video_dir)
all_sess = [
    d for d in os.listdir(root_video_dir)
    if os.path.isdir(os.path.join(root_video_dir, d))
]
train_sess = all_sess 

# DataLoader for training
train_dl = DataLoader(
    PTSDDataset(
        train_e,
        train_sess,
        root_video_dir,
        train_audio_dir,
        train_text_dir
    ),
    batch_size=BATCH_SIZE,
    shuffle=True,
    num_workers=4,
    pin_memory=True
)

# DataLoader for validation (sessions = ["."],   val_session_dir/<CLASS> )
val_dl = DataLoader(
    PTSDDataset(
        val_e,
        sessions=["."],     
        vroot=val_session_dir,
        aroot=val_audio_dir,
        troot=val_text_dir
    ),
    batch_size=BATCH_SIZE,
    shuffle=False,
    num_workers=4,
    pin_memory=True
)

# (5.1) Compute and print dataset statistics before training
import os

train_video_dir = root_video_dir
val_video_dir   = val_session_dir

# Count video frames in training and validation sets
total_train_video_frames = 0
total_val_video_frames = 0
for cls in class_names:
    tvc = os.path.join(train_video_dir, cls)
    if os.path.isdir(tvc):
        total_train_video_frames += len([f for f in os.listdir(tvc) 
                                         if f.lower().endswith(('.png','.jpg','.jpeg'))])
    vvc = os.path.join(val_video_dir, cls)
    if os.path.isdir(vvc):
        total_val_video_frames   += len([f for f in os.listdir(vvc) 
                                         if f.lower().endswith(('.png','.jpg','.jpeg'))])

print(f"Total video frames in training set:   {total_train_video_frames}")
print(f"Total video frames in validation set: {total_val_video_frames}")

# Count audio spectrogram PNG files
total_train_audio_files = 0
total_val_audio_files   = 0
for cls in class_names:
    tad = os.path.join(train_audio_dir, cls)
    if os.path.isdir(tad):
        total_train_audio_files += len([f for f in os.listdir(tad) if f.lower().endswith('.png')])
    vad = os.path.join(val_audio_dir, cls)
    if os.path.isdir(vad):
        total_val_audio_files   += len([f for f in os.listdir(vad) if f.lower().endswith('.png')])

print(f"Total audio spectrograms in training set:   {total_train_audio_files}")
print(f"Total audio spectrograms in validation set: {total_val_audio_files}")

# Count text samples
total_train_text_files = 0
total_val_text_files   = 0
for cls in class_names:
    ttd = os.path.join(train_text_dir, cls)
    if os.path.isdir(ttd):
        total_train_text_files += len([f for f in os.listdir(ttd) if f.endswith('.txt')])
    vtd = os.path.join(val_text_dir, cls)
    if os.path.isdir(vtd):
        total_val_text_files   += len([f for f in os.listdir(vtd) if f.endswith('.txt')])

print(f"Total text samples in training set:   {total_train_text_files}")
print(f"Total text samples in validation set: {total_val_text_files}")

# Count number of patients (unique IDs) per modality
num_train_video_patients = len(train_e)
num_val_video_patients   = len(val_e)
num_train_audio_patients = len(train_e)
num_val_audio_patients   = len(val_e)
num_train_text_patients  = len(train_e)
num_val_text_patients    = len(val_e)

print(f"Video patients - train: {num_train_video_patients}, val: {num_val_video_patients}")
print(f"Audio patients - train: {num_train_audio_patients}, val: {num_val_audio_patients}")
print(f"Text  patients - train: {num_train_text_patients}, val: {num_val_text_patients}")

class PTSDVideoTransformer(nn.Module):
    """
    Video branch: tubelet conv ‚Üí pool ‚Üí flatten ‚Üí Linear(256 ‚Üí NUM_CLASSES).
    We freeze conv backbone; train only the projection layer.
    Output: (B, NUM_CLASSES) raw scores (no sigmoid).
    """
    def __init__(self, num_classes=NUM_CLASSES):
        super().__init__()
        # Tubelet embedding conv: projects 16√ó16 patches across frames into 256-dim features
        self.conv = nn.Conv3d(
            in_channels=3,
            out_channels=256,
            kernel_size=(1, 16, 16),
            stride=(1, 16, 16)
        )
        self.pool = nn.AdaptiveAvgPool3d((1, 1, 1))
        self.flatten = nn.Flatten()
        # Projection head to map 256-dim embedding to NUM_CLASSES logits
        self.proj = nn.Linear(256, num_classes)

    def forward(self, x: torch.Tensor) -> torch.Tensor:
        """
        x: (B, 3, SEQ_LEN, IMG_H, IMG_W)
        Returns: (B, NUM_CLASSES) raw logits
        """
        feat = self.conv(x)                  # ‚Üí (B, 256, T', H', W')
        feat = self.pool(feat)               # ‚Üí (B, 256, 1, 1, 1)
        feat = self.flatten(feat)            # ‚Üí (B, 256)
        return self.proj(feat)               # ‚Üí (B, NUM_CLASSES)

    def extract_feature(self, x: torch.Tensor) -> torch.Tensor:
        """
        Extract 256-dim embedding before projection.
        Output: (B, 256)
        """
        feat = self.conv(x)
        feat = self.pool(feat)
        feat = self.flatten(feat)            # ‚Üí (B, 256)
        return feat

# --- Load checkpoint, initialize, and setup ---
ckpt = torch.load(ckpt_video, map_location='cpu', weights_only=True)
if isinstance(ckpt, dict) and 'model' in ckpt:
    ckpt = ckpt['model']
# Strip prefixes for conv backbone
ckpt = strip_prefix(ckpt, 'module.')
ckpt = strip_prefix(ckpt, 'backbone.')

video_model = PTSDVideoTransformer(num_classes=NUM_CLASSES).to(device)
# Load pretrained conv weights; projector will be randomly initialized
missing_keys, unexpected_keys = video_model.load_state_dict(ckpt, strict=False)
print(f"‚úÖ Loaded conv backbone weights: {len(missing_keys)} missing, {len(unexpected_keys)} unexpected keys.")

# Freeze all conv backbone params; train only projection head
for name, param in video_model.named_parameters():
    if name.startswith('proj'):
        param.requires_grad = True
    else:
        param.requires_grad = False

# Set model to train mode for fine-tuning projection
video_model.train()
# (7) audio EfficientNet‚ÄëV2‚ÄëL + text BERT

# ‚Äë‚Äë audio
audio_model = models.efficientnet_v2_l(weights=None)
audio_model.classifier[-1] = nn.Linear(audio_model.classifier[-1].in_features, NUM_CLASSES)
aud_sd = strip_prefix(torch.load(ckpt_audio, map_location='cpu'),'module.')
audio_model.load_state_dict(aud_sd, strict=False)
for p in audio_model.parameters(): p.requires_grad=False
audio_model.to(device).eval()

# ‚Äë‚Äë text
# 1) Allow XGBClassifier in torch.load‚Äôs safe globals
torch.serialization.add_safe_globals([XGBClassifier])

# 1) First try loading with joblib (in case it was saved that way)
try:
    ckpt_data = joblib.load(ckpt_text)
    # If it‚Äôs a dict containing a 'state_dict' key, extract it
    if isinstance(ckpt_data, dict) and 'state_dict' in ckpt_data:
        raw_state_dict = ckpt_data['state_dict']
    else:
        # Otherwise assume the dict itself is the state_dict
        raw_state_dict = ckpt_data
    print("‚úÖ Checkpoint loaded with joblib")
except Exception:
    # Fallback: load with torch.load
    raw = torch.load(ckpt_text, map_location='cpu')
    if isinstance(raw, dict) and 'state_dict' in raw:
        raw_state_dict = raw['state_dict']
    else:
        raw_state_dict = raw
    print("‚úÖ Checkpoint loaded with torch.load")

# 2) Strip any 'module.' prefixes (if checkpoint was saved from a DataParallel/DistributedDataParallel model)
def strip_prefix(state_dict, prefix='module.'):
    return {
        (k[len(prefix):] if k.startswith(prefix) else k): v
        for k, v in state_dict.items()
    }

state_dict = strip_prefix(raw_state_dict, 'module.')

# 3) Initialize the BERT model exactly as in training
text_model = BertForSequenceClassification.from_pretrained(
    'bert-base-uncased',
    num_labels=NUM_CLASSES
)

# 4) Load the weights into the model (non-strict to allow slight mismatches)
text_model.load_state_dict(state_dict, strict=False)
print("‚úÖ State dict loaded into the model")

# 5) Freeze all layers except the top two encoder layers and the classifier head
for name, param in text_model.named_parameters():
    if 'encoder.layer.10' in name or 'encoder.layer.11' in name or 'classifier' in name:
        param.requires_grad = True
    else:
        param.requires_grad = False

# Move to device and set to train mode
text_model.to(device).train()
print("‚úÖ Model layers frozen/unfrozen, ready for fusion")
# (8) fusion network + loss + scheduler

class FusionHead(nn.Module):
    def __init__(self,nc):
        super().__init__()
        self.wv = self.wa = self.wt = nn.Parameter(torch.ones(1))
        self.norm = nn.LayerNorm(nc*3)
        self.dp   = nn.Dropout(0.3)
        self.fc   = nn.Linear(nc*3, nc)
    def forward(self,v,t,a):
        x = torch.cat([self.wv*v, self.wt*t, self.wa*a], 1)
        return self.fc(self.dp(self.norm(x)))
# (8) fusion network + loss + scheduler

# FusionHead waise hi rahegi (no change)

class LateFusion(nn.Module):
    def __init__(self, vm, tm, am, nc):
        super().__init__()
        self.vm, self.tm, self.am = vm, tm, am
        self.head = FusionHead(nc)

    def forward(self, vids, auds, ids, mask):
        # Video branch
        v = self.vm(vids)  # [B, nc]

        # Audio branch: per-frame 2D processing + pooling
        B, C, T, H, W = auds.shape
        x = auds.permute(0, 2, 1, 3, 4).reshape(B * T, C, H, W)
        fl = self.am(x)                # [B*T, nc]
        a  = fl.view(B, T, -1).mean(dim=1)  # [B, nc]

        # Text branch
        t = self.tm(input_ids=ids, attention_mask=mask).logits  # [B, nc]

        # Fusion & final classifier
        return self.head(v, t, a)

model = LateFusion(video_model, text_model, audio_model, NUM_CLASSES).to(device)

optimizer = AdamW(
    [p for p in model.parameters() if p.requires_grad],
    lr=LR, weight_decay=1e-2
)
total_steps = len(train_dl) * EPOCHS
scheduler = get_cosine_schedule_with_warmup(
    optimizer,
    total_steps // 20,
    total_steps
)
criterion = nn.CrossEntropyLoss(weight=class_wts)


# (9) train / validate with tqdm for both training and validation

best_val_acc = 0
train_hist, val_hist = [], []

for ep in range(1, EPOCHS + 1):
    # --- TRAINING ---
    model.train()
    tl, tc, tn = 0.0, 0, 0
    for v, a, ids, mask, y in tqdm(train_dl, desc=f"Epoch {ep}/{EPOCHS} - Training"):
        v, a, ids, mask, y = v.to(device), a.to(device), ids.to(device), mask.to(device), y.to(device)
        optimizer.zero_grad()
        out = model(v, a, ids, mask)
        loss = criterion(out, y)
        loss.backward()
        torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)
        optimizer.step()
        scheduler.step()

        tl += loss.item() * y.size(0)
        preds = out.argmax(1)
        tc += (preds == y).sum().item()
        tn += y.size(0)

    train_loss = tl / tn
    train_acc  = 100 * tc / tn
    train_hist.append((train_loss, train_acc))

    # --- VALIDATION ---
    model.eval()
    vl, vc, vn = 0.0, 0, 0
    for v, a, ids, mask, y in tqdm(val_dl, desc=f"Epoch {ep}/{EPOCHS} - Validation"):
        v, a, ids, mask, y = v.to(device), a.to(device), ids.to(device), mask.to(device), y.to(device)
        out = model(v, a, ids, mask)
        loss = criterion(out, y)

        vl += loss.item() * y.size(0)
        preds = out.argmax(1)
        vc += (preds == y).sum().item()
        vn += y.size(0)

    val_loss = vl / vn
    val_acc  = 100 * vc / vn
    val_hist.append((val_loss, val_acc))

    # Checkpoint
    if val_acc > best_val_acc:
        best_val_acc = val_acc
        torch.save(model.state_dict(), 'best_fusion_model.pth')
        print(f"  üî• New best! Saved at epoch {ep}")

    # Epoch summary
    print(f"Epoch {ep:02d} | "
          f"Train Loss={train_loss:.4f}, Train Acc={train_acc:.2f}%  | "
          f"Val Loss={val_loss:.4f}, Val Acc={val_acc:.2f}%")

# (10) loss & accuracy curve plot kare‚Äëga

import matplotlib.pyplot as plt

# unpack
tr_loss, tr_acc = zip(*train_hist)
vl_loss, vl_acc = zip(*val_hist)
epochs = range(1, EPOCHS+1)

plt.figure(figsize=(6,4))
plt.plot(epochs, tr_loss, label='Train Loss')
plt.plot(epochs, vl_loss, label='Val Loss')
plt.xlabel('Epoch'); plt.ylabel('Loss'); plt.title('Loss Curve'); plt.legend(); plt.grid()
plt.show()

plt.figure(figsize=(6,4))
plt.plot(epochs, tr_acc, label='Train Acc')
plt.plot(epochs, vl_acc, label='Val Acc')
plt.xlabel('Epoch'); plt.ylabel('Accuracy (%)'); plt.title('Accuracy Curve'); plt.legend(); plt.grid()
plt.show()

print(f'üê±‚ÄçüèÜ  Best validation accuracy: {best_val_acc:.2f}%')

from IPython.display import FileLink
FileLink('best_fusion_model.pth')
```
