```python
# EfficientNet + ViT Ensemble for PTSD Spectrogram Classification

import sys, os
import torch
import torch.nn as nn
import torch.optim as optim
import torchvision
import torchvision.transforms as transforms
from torch.utils.data import DataLoader, Dataset
from tqdm import tqdm
import matplotlib.pyplot as plt
import numpy as np

# Device configuration
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using device:", DEVICE)

# Hyperparameters
BATCH_SIZE = 16
EPOCHS_FINAL = 20
NUM_CLASSES = 2
BEST_LR = 3e-4
BEST_DROPOUT = 0.4
BEST_WEIGHT_DECAY = 1e-5
MIXUP_ALPHA = 0.5

# Dataset paths
audio_train_dir = "/kaggle/input/ptsd-multimodel-dataset/sorted_multimodal/audio/training"
audio_val_dir   = "/kaggle/input/ptsd-multimodel-dataset/sorted_multimodal/audio/validation"

# SpecAugment for spectrograms
class SpecAugment(object):
    def __init__(self, freq_mask_param=15, time_mask_param=20, num_freq_masks=1, num_time_masks=1):
        self.freq_mask_param = freq_mask_param
        self.time_mask_param = time_mask_param
        self.num_freq_masks = num_freq_masks
        self.num_time_masks = num_time_masks

    def __call__(self, spec):
        if spec.dim() == 2:
            spec = spec.unsqueeze(0)
        augmented_spec = spec.clone()
        C, H, W = augmented_spec.shape
        for _ in range(self.num_freq_masks):
            f = torch.randint(0, self.freq_mask_param, (1,)).item()
            f0 = torch.randint(0, max(H - f, 1), (1,)).item()
            augmented_spec[:, f0:f0+f, :] = 0
        for _ in range(self.num_time_masks):
            t = torch.randint(0, self.time_mask_param, (1,)).item()
            t0 = torch.randint(0, max(W - t, 1), (1,)).item()
            augmented_spec[:, :, t0:t0+t] = 0
        return augmented_spec

# Transforms
transform_train = transforms.Compose([
    transforms.Grayscale(num_output_channels=3),
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    SpecAugment(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406],
                         std=[0.229, 0.224, 0.225])
])

transform_val = transforms.Compose([
    transforms.Grayscale(num_output_channels=3),
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406],
                         std=[0.229, 0.224, 0.225])
])

# Dataset and DataLoader
class AudioImageDataset(Dataset):
    def __init__(self, root, transform=None):
        self.dataset = torchvision.datasets.ImageFolder(root, transform=transform)
    def __len__(self):
        return len(self.dataset)
    def __getitem__(self, idx):
        image, label = self.dataset[idx]
        return image, label

train_dataset = AudioImageDataset(audio_train_dir, transform=transform_train)
val_dataset   = AudioImageDataset(audio_val_dir, transform=transform_val)

train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True,
                          num_workers=4, pin_memory=True, drop_last=True)
val_loader   = DataLoader(val_dataset,   batch_size=BATCH_SIZE, shuffle=False,
                          num_workers=4, pin_memory=True, drop_last=False)

print(f"Train Samples: {len(train_dataset)} | Validation Samples: {len(val_dataset)}")

# Ensemble Model: EfficientNetV2_L + ViT_B_16
class PTSDClassifierEnsemble(nn.Module):
    def __init__(self, num_classes=2, dropout_rate=0.4):
        super(PTSDClassifierEnsemble, self).__init__()

        # EfficientNetV2-L
        self.effnet = torch.hub.load('pytorch/vision', 'efficientnet_v2_l', weights='EfficientNet_V2_L_Weights.DEFAULT')
        self.effnet.classifier = nn.Identity()  # Remove classifier
        self.effnet_out = 1280  # Output features of efficientnet_v2_l

        # ViT_B_16
        self.vit = torchvision.models.vit_b_16(pretrained=True)
        self.vit.heads = nn.Identity()  # Remove classifier
        self.vit_out = 768  # Output features of vit_b_16

        # Combined classifier
        self.dropout = nn.Dropout(p=dropout_rate)
        self.classifier = nn.Linear(self.effnet_out + self.vit_out, num_classes)

    def forward(self, x):
        x1 = self.effnet(x)  # EfficientNet features
        x2 = self.vit(x)     # ViT features
        x_cat = torch.cat((x1, x2), dim=1)
        x_cat = self.dropout(x_cat)
        return self.classifier(x_cat)

final_model = PTSDClassifierEnsemble(num_classes=NUM_CLASSES, dropout_rate=BEST_DROPOUT).to(DEVICE)

# Loss, Optimizer, Scheduler
criterion = nn.CrossEntropyLoss()
optimizer = optim.AdamW(final_model.parameters(), lr=BEST_LR, weight_decay=BEST_WEIGHT_DECAY)
scheduler = torch.optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=EPOCHS_FINAL)

# Mixup function
def mixup_data(x, y, alpha=MIXUP_ALPHA):
    if alpha > 0:
        lam = np.random.beta(alpha, alpha)
    else:
        lam = 1
    batch_size = x.size()[0]
    index = torch.randperm(batch_size).to(x.device)
    mixed_x = lam * x + (1 - lam) * x[index, :]
    y_a, y_b = y, y[index]
    return mixed_x, y_a, y_b, lam

# Training loop
best_val_acc = 0.0
best_model_epoch = 0
best_model_path = "best_effnet_vit_ensemble.pth"

train_losses, val_losses = [], []
train_accuracies, val_accuracies = [], []

for epoch in range(EPOCHS_FINAL):
    final_model.train()
    running_loss = 0.0
    correct_train = 0
    total_train = 0

    pbar_train = tqdm(train_loader, desc=f"Epoch {epoch+1}/{EPOCHS_FINAL} [Train]")
    for images, labels in pbar_train:
        images, labels = images.to(DEVICE), labels.to(DEVICE)
        mixed_images, targets_a, targets_b, lam = mixup_data(images, labels, alpha=MIXUP_ALPHA)
        optimizer.zero_grad()
        outputs = final_model(mixed_images)
        loss = lam * criterion(outputs, targets_a) + (1 - lam) * criterion(outputs, targets_b)
        loss.backward()
        optimizer.step()

        running_loss += loss.item()
        _, predicted = torch.max(outputs, dim=1)
        correct_train += (lam * (predicted == targets_a).sum().item() +
                          (1 - lam) * (predicted == targets_b).sum().item())
        total_train += labels.size(0)
        pbar_train.set_postfix({"loss": loss.item()})

    train_loss_epoch = running_loss / len(train_loader)
    train_acc_epoch = 100.0 * correct_train / total_train
    train_losses.append(train_loss_epoch)
    train_accuracies.append(train_acc_epoch)

    final_model.eval()
    val_running_loss = 0.0
    correct_val = 0
    total_val = 0

    pbar_val = tqdm(val_loader, desc=f"Epoch {epoch+1}/{EPOCHS_FINAL} [Val]")
    with torch.no_grad():
        for images, labels in pbar_val:
            images, labels = images.to(DEVICE), labels.to(DEVICE)
            outputs = final_model(images)
            loss = criterion(outputs, labels)
            val_running_loss += loss.item()
            _, predicted = torch.max(outputs, dim=1)
            correct_val += (predicted == labels).sum().item()
            total_val += labels.size(0)

    val_loss_epoch = val_running_loss / len(val_loader)
    val_acc_epoch = 100.0 * correct_val / total_val
    val_losses.append(val_loss_epoch)
    val_accuracies.append(val_acc_epoch)
    scheduler.step()

    print(f"Epoch [{epoch+1}/{EPOCHS_FINAL}] "
          f"Train Loss: {train_loss_epoch:.4f}, Train Acc: {train_acc_epoch:.2f}% "
          f"| Val Loss: {val_loss_epoch:.4f}, Val Acc: {val_acc_epoch:.2f}%")

    if val_acc_epoch > best_val_acc:
        best_val_acc = val_acc_epoch
        best_model_epoch = epoch + 1
        torch.save(final_model.state_dict(), best_model_path)
        print(f"‚úÖ Best model saved at epoch {best_model_epoch} with Val Acc: {best_val_acc:.2f}%\n")

print("\nüéâ Training Completed!")
print(f"üèÜ Best Validation Accuracy: {best_val_acc:.2f}% at Epoch {best_model_epoch}")

# Plot Loss and Accuracy
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(range(1, EPOCHS_FINAL + 1), train_losses, label='Train Loss')
plt.plot(range(1, EPOCHS_FINAL + 1), val_losses, label='Val Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Loss Curve')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(range(1, EPOCHS_FINAL + 1), train_accuracies, label='Train Acc')
plt.plot(range(1, EPOCHS_FINAL + 1), val_accuracies, label='Val Acc')
plt.xlabel('Epoch')
plt.ylabel('Accuracy (%)')
plt.title('Accuracy Curve')
plt.legend()

plt.show()
```
